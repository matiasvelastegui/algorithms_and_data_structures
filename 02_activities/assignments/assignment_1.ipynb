
-   Code the solution to your assigned problem in Python (code chunk). Note: each problem can be solved more simply if you use an abstract data type that is suitable for that problem. Using that try to find the best time and space complexity solution!


````python
from typing import List

def first_duplicate(nums: List[int]) -> int:
    seen = set()
    for num in nums:
        if num in seen:
            return num
        seen.add(num)
    return -1

# Example usage:
print(first_duplicate([8, 8, 11, 100]))  # Output: 8
print(first_duplicate([100, 200, 300, 300, 100]))  # Output: 300
print(first_duplicate([1, 2, 3, 4, 5]))  # Output: -1
````


-   Explain why your solution works


````python
# The function `first_duplicate` iterates through the list of integers, keeping track of seen numbers in a set. Using a set allows so that if it encounters a number that has already been seen, it returns that number as the first duplicate.
````


-   Explain the problem’s time and space complexity


````python
# The time complexity of the optimal solution is O(n), where n is the length of the input list. This is because we need to check each element in the list once to determine if it has already been seen.

# The space complexity is also O(n) in the worst case, because we may need to store up to n elements in a set if there are no duplicates in the list.
````


-   Explain the thinking to an alternative solution (no coding required, but a classmate reading this should be able to code it up based off your text)


````python
# An alternative solution is to use a brute-force approach: for each element in the list, check all previous elements to see if it has already appeared.
# If a duplicate is found, return it immediately. If no duplicates are found after checking all elements, return -1.

def first_duplicate_alt(nums: List[int]) -> int:
    for i in range(1, len(nums)):
        for j in range(i):
            if nums[i] == nums[j]:
                return nums[i]
    return -1

# This method has O(n^2) time complexity and O(1) space complexity (ignoring input storage).
````
