  # Question Three: Missing Number in Range
 
  You are given a list containing `n` integers in the range `[0, n]`. Return a list of numbers that are missing from the range `[0, n]` of the array. If there is no missing number, return -1. Note, all the integers in the list may not be unique.
  
  ## Examples

  ### Example 1

  Input: `lst = [0, 2]`

  Output: [1]

  ### Example 2

  Input: `lst = [5, 0, 1]`

  Output: [2, 3, 4]

  ### Example 3

  Input: `lst = [6, 8, 2, 3, 5, 7, 0, 1, 10]`

  Output: [4, 9]

</details>

#### Starter Code for Question 3


```python
from typing import List

def missing_num(nums: list) -> list:
    n = max(nums) if nums else 0
    full_set = set(range(0, n + 1))
    nums_set = set(nums)
    missing = sorted(list(full_set - nums_set))
    return missing if missing else -1

print(missing_num([0, 2])) # should return 1
print(missing_num([5, 0, 1])) # should return 2, 3, 4
print(missing_num([6, 8, 2, 3, 5, 7, 0, 1, 10])) # should return 4, 9
print(missing_num([0, 1, 2, 3, 4, 5])) # should return -1
```


-   Copy the solution your partner wrote. 


```python
def missing_num(nums: List) -> int:
    n = max(nums)
    nums_set = set(nums)
    missing = [i for i in range(n+1) if i not in nums_set] # find missing numbers and store in a list
    return missing if missing else -1

# Testing the missing_num function
print(missing_num([0, 2]))
print(missing_num([5, 0, 1]))
print(missing_num([6, 8, 2, 3, 5, 7, 0, 1, 10]))
print(missing_num([0, 1, 2]))
print(missing_num([0, 2, 2, 2]))
```


-   Explain why their solution works in your own words.


```python
# The function uses a clever choice of data structures to efficiently find missing numbers, which is a set for O(1) lookups.
# It iterates through the range from 0 to the maximum number in the list and checks if each number is present in the set of numbers from the list.
# If a number is not found, it is added to the list of missing numbers.
# if there are no missing numbers, it returns -1, otherwise it returns the list of missing numbers.
```
